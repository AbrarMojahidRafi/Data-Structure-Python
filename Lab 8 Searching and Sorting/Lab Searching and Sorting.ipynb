```python
# Task 1 Selection sort, array

def selectionSortArray(arr):
  for i in range(len(arr)):
    minPos=0
    for j in range(i,len(arr)):
      if arr[j] < arr[i]:
        minPos=j
        print(minPos)
    temp=arr[i] 
    
    arr[i]=arr[minPos] 
    arr[minPos]=temp 

  return arr


arr=[9,8,7,6,5,4,3,2,1]
selectionSortArray(arr)
```
