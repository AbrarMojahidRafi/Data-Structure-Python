```python
class KeyIndex:
  def __init__(self,a):
    self.arr=None
    self.positiveOrNegative = None
    self.p=None

    allElementIsPositive=True
    for i in a:
      if i<0:
        allElementIsPositive=False 

    if allElementIsPositive == True:
      self.positiveOrNegative="positive"
      max=0
      for i in a:
        if i > max:
          max=i
      auxArr=[0]*(max+1)
      for i in a:
        auxArr[i]+=1
      self.arr=auxArr

    else:
      self.positiveOrNegative="negative"
      max=0
      for i in a:
        if i > max:
          max=i
      
      min=max
      for i in a:
        if i<min:
          min=i

      p=(-1)*min
      self.p=p

      auxArr=[0]*(max+p+1)
      for i in a:
        auxArr[i+p] += 1
      self.arr=auxArr
  
  def search(self, val):
    if self.arr[val]!=0:
      return True
    else:
      return False 

  def sort(self):
    if self.positiveOrNegative=="positive":
      size=0
      for i in self.arr:
        if i!=0:
          size+=i

      l=[0]*(size)
      idx=0
      for i in range(len(self.arr)):
        for j in range(self.arr[i]):
          l[idx]=i 
          idx+=1
      return l

    else:
      size=0
      for i in self.arr:
        if i!=0:
          size+=i

      l=[0]*(size)
      idx=0
      for i in range(len(self.arr)):
        for j in range(self.arr[i]):
          l[idx]=i-self.p
          idx+=1
      return l


k1=KeyIndex([1,9,8,-45,10,5,3,1])
print(k1.search(0))
print(k1.sort())
```

```python
def Hashtable(a):
  hashTable=[0]*9
  for i in a:
    consonant = 0
    digitSum = 0
    for j in i:
      if (j not in "AEIOU") and (j not in "0123456789"):
        consonant += 1 
      elif (j in "0123456789"):
        digitSum += int(j)

    idx = (consonant*24 + digitSum) % 9 
    
    if hashTable[idx] ==0:
      hashTable[idx] = i
    else:
      cirIdx = (idx + 1)% 9
      while cirIdx != idx:
        if hashTable[cirIdx] == 0:
          hashTable[cirIdx] = i
          break
        cirIdx = (cirIdx + 1)%9

  return hashTable

a=['ST1E89B8A32', "MOJAHIDE89B8A32",'ST189B832','ST189B832','RAFT1165','ABRAR1953']
# a=['ST1E89B8A32', "ST1E89B8A32"]
Hashtable(a)
```
