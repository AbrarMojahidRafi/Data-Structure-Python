```python
#################################################################
########################### Task 01 #############################
#################################################################

class ArrayStack:
  def __init__(self,size):
    self.stack=[0]*size
    self.idx=-1
  def push(self, value):
    if self.idx==len(self.stack)-1:
      return ("Stack is full")
    else:
      self.idx += 1
      self.stack[self.idx]=value
  def pop(self):
    if self.idx==-1:
      return ("Stack is empty")
    else:
      temp=self.stack[self.idx]
      self.stack[self.idx] = 0
      self.idx -= 1
      return temp
  def peek(self):
    if self.idx==-1:
      return ("Stack is empty")
    else:
      return self.stack[self.idx]
```
